# **¿Qué buscamos hacer con nuestra red neuronal?**

Por el momento, lo que buscaremos es trazar líneas, o manipular la imagen de tal manera que podamos separar los puntos azules y naranjas de la mejor manera. Para ello podemos hacer uso de herramientas, o funciones activadoras como:

1. Funciones líenales: Las cuales básicamente son lienas rectas que se pueden trazar en el plano para lograr separar los puntos.
2. Funciones tangenciales : Las cuales dejan de ser lineas rectas y toman formas curvas. ¿Ahora tienes más posibilidades para separar los puntos, no?
3. Funciones sigmoides: Estas funciones son buenas para lograr separar los puntos en solo dos espacios, donde uno sea extrictamente azul y el otro extrictamente azul.
4. Función ReLU: Actualemente, esta es la función o método que más se usa, ya que básicamente anula los datos que no queremos (sean naranjas o azules) y nos deja los datos que queremos. Es una muy buena manera de poder separar los puntos de su contraparte.
